```python
# importação das bibliotecas
import cv2
import numpy as npy
from matplotlib import pyplot as plt
# leitura da imagem
img=cv2.imread("colorida.jpg")

#metodo para cortar a imagem em RGB (importante seguir a ordem de vermelho,verde e azul pois caso contrario mudará o resultado)
vermelho, verde, azul = cv2.split(img)
# equalizando cada um dos cortes RGB
vermelhoEqualizada= cv2.equalizeHist(vermelho)
verdeEqualizada= cv2.equalizeHist(verde)
azulEqualizada= cv2.equalizeHist(azul)
#definindo o tamanho da figura de saida
fig= plt.figure(figsize=(20,5))
#localizando onde vai sair (linha|colunatotais|coluna localizada)
ax1= fig.add_subplot(191)
#imprimindo o historiograma da cor azul com o valor maximo de 256 variando de 0 a 256
plt.hist(azul.ravel(),256,[0,256])
#titulo do historiograma 
plt.title("Azul")

#repetindo o processo acima para o verde
ax2= fig.add_subplot(192)
plt.hist(verde.ravel(),256,[0,256])
plt.title("Verde")

#repetindo o processo acima para o vermelho
ax3= fig.add_subplot(193)
plt.hist(vermelho.ravel(),256,[0,256])
plt.title("Vermelho")

#mudando o tamanho da janela popup(imagem, parametro da img de saida, fator de escala horizontal(x), fator de escala vertical(y),tipo de interpolação(cubica))
imgpequena= cv2.resize(img,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_CUBIC)

#imprimindo o popup (nome da imagem, variavel que foi feito o resize)
cv2.imshow("imagem original", imgpequena)

#repetindo o processo de imprimir o historiograma para o azulEqualizado
ax1_1= fig.add_subplot(196)
plt.hist(azulEqualizada.ravel(),256,[0,256])
plt.title("Azul equalizada")

#repetindo o processo de imprimir o historiograma para o verdeEqualizado
ax2_2= fig.add_subplot(197)
plt.hist(verdeEqualizada.ravel(),256,[0,256])
plt.title("Verde equalizada")

#repetindo o processo de imprimir o historiograma para o vermelhoEqualizado
ax3_2= fig.add_subplot(198)
plt.hist(vermelhoEqualizada.ravel(),256,[0,256])
plt.title("Vermelho equalizada")

#juntando o valor das equalizações feitas acima
imgEqualizada = cv2.merge((vermelhoEqualizada , verdeEqualizada , azulEqualizada))

#repetindo o processo de resize para a img equalizada
imgEqualizadinha= cv2.resize(imgEqualizada,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_CUBIC)

#repetindo o processo de imprimir a img para a img equalizadinha
cv2.imshow("imagem equalizada",imgEqualizadinha)

#mostrando a plt
plt.show()

#esperando que o usuario clique alguma tecla para fechar a imagem
cv2.waitKey(0)

#quando clickar vai destroir todas as janelas popups criadas
cv2.destroyAllWindows()




```
